```python
# we will start simple with a single fully-connected neural layer as encoder and decoder

ENCODING_DIM = 2500
HIDDEN_LAYER_A_DIM = 9300
HIDDEN_LAYER_B_DIM = 16200
INPUT_DIM = total_feature




input_layer = tf.keras.layers.Input(shape=(INPUT_DIM,))


encoder = tf.keras.layers.Dense(ENCODING_DIM, activation='relu')(input_layer)


decoded = tf.keras.layers.Dense(INPUT_DIM, activation='sigmoid')(encoder)


autoencoder = tf.keras.models.Model(input_layer, decoded)
```

```python
autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
```

```python
autoencoder.fit(input_data_t, input_data_t, epochs=5, batch_size=8, shuffle=True, validation_data=(val_data_t, val_data_t))
```
